### INPUT_FOLDER structure:
- INPUT_FOLDER should has structure as same as requirement of organizer, contain "test.json" as follow:\

├── test.json \
├── test \
│   ├── 0.jpg \
│   ├── 100.png \
│   ├── 101.jpg \
│   ├── 102.jpg \
- With "test/0.jpg" is an "img_local_path" in "test.json" for example.

- DATA: absolutely path to test.json file of organizer
